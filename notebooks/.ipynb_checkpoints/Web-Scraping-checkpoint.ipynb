```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10
page_size = 100

url = f"{base_url}/page/1/?sortby=post_date%3ADesc&pagesize={page_size}"
```

```python
#Collect HTML data from this page
response = requests.get(url)
response
```

```python
response.text
```

```python
#Get the content of the response sent by the server
content = response.content
content
```

```python
#Print the headers sent by the server
print(response.headers)
```

> Note: Youll want to pass page.content instead of page.text to avoid problems with character encoding. The .content attribute holds raw bytes, which can be decoded better than the text representation you printed earlier using the .text attribute.

```python
parsed_content = BeautifulSoup(content, 'html.parser')
parsed_content
```

```python
#We can either use {"class": "text_content"} as in this example, or class_="text_content" as in the next example to get this class
result = parsed_content.find_all("div", {"class": "text_content"})
result
```

```python
result2 = parsed_content.find_all("div", class_="text_content")
result2
```

```python
#Find the text within each element
texts = []
for item in result:
    text_result = item.text.strip()
    texts.append(text_result)
```

```python
texts[:2]
```

```python
#Get the parent <div> of the current element
parents = [element.parent for element in result]
parents[:1]
```

```python
#Get the URL from the attribute "href" in element <a>
for element in result:
    # links will have all <a> elements 
    links = element.find_all("a")
    for link in links:
        link_url = link["href"]
        print(f"Link is: {link_url}\n")
```
